# Portfolio C
### CASE: Buy some stocks and short some other stocks.  There should be 3 or more stocks in the portfolio

For this case, we will buy some stocks and short some stocks for making a portfolio. 

We have the following **information** about the stocks ['AAPL', 'AMZN', 'NFLX'. 'META', 'GOOGL']:
* **average return** of the stocks, vector $\vec\mu$
* **standard deviation** of the stocks, vector $\vec\sigma$
* **correlation** between the stocks, matrix  $R$



As we are talking about more than one stock for calculating the portfolio statistics, we will use Vectorized calculations for calculating the portfolio statistics. For a given average return vector, standard deviation vector and correlation matrix, we can calculate the portfolio statistics using the following formulae:

1. Portfolio mean/average is given by \
$ \mu_{p} = \sum\limits _{i=1} ^N \mu_i * w_i$ = $\vec\mu^T * \vec w$ = $\vec \mu \odot \vec w$

2. Given a Covariance Matrix *C*, Portfolio variance  can be calcaulted using the following formula: Quad form (W^T * *C* * W)\
$ \sigma_p^2= {\vec w^T C \vec w}$

But, we need to compute the **covariance matrix** $C$ first, given the correlation matrix $R$ and the standard deviation vector. The covariance matrix is given by:

* for two stocks, $C_{ij} = \sigma_i * \sigma_j * \rho_{ij}$

* Vectorized form $C = \mathbf{R} * (\vec\sigma \otimes \vec\sigma)$ **or**  $\mathbf{\sigma} = \mathbf{I}\vec\sigma$ followed by , $C = \sigma \mathbf{R} \sigma)$


where $\vec\sigma$ is the standard deviation vector and $R$ is the correlation matrix.
